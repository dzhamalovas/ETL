# Отчет по категориям (Доп. задание 1)

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(x="category", y="total_profit", data=category_report)
plt.title("Общая прибыль по категориям (Profit > 0)")
plt.xlabel("Категория")
plt.ylabel("Total Profit")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="category", y="total_sales", data=category_report)
plt.title("Общий объем продаж по категориям")
plt.xlabel("Категория")
plt.ylabel("Total Sales")
plt.xticks(rotation=30)
plt.show()
```

# Анализ доставки (Доп. задание 2)

```python
plt.figure(figsize=(8,5))
sns.barplot(x="ship_mode", y="total_profit", data=shipping_analysis)
plt.title("Общая прибыль по способу доставки")
plt.xlabel("Способ доставки")
plt.ylabel("Total Profit")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="ship_mode", y="avg_profit", data=shipping_analysis)
plt.title("Средняя прибыль по способу доставки")
plt.xlabel("Способ доставки")
plt.ylabel("Average Profit")
plt.xticks(rotation=30)
plt.show()
```
